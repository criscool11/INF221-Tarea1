```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data.csv'
data = pd.read_csv(file_path)

data_clean = data.dropna(subset=['Age', 'Fav genre', 'Music effects'])
data_clean['Age'] = pd.to_numeric(data_clean['Age'], errors='coerce')

analysis_1 = data_clean.groupby(['Music effects', 'Fav genre'])['Age'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(y='Age', x='Fav genre', hue = 'Music effects', palette='Set1', data=analysis_1)
plt.title('Promedio de edad en funcion del genero favorito y estratificada por los efectos sobre salud mental', fontsize=14)
plt.xlabel('Generos Musicales', fontsize=12)
plt.ylabel('Edad Promedio', fontsize=12)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=analysis_1, x='Fav genre', y='Age', hue='Music effects', s=100, palette='Set1')
plt.title('Relación entre Género Favorito y Edad Promedio Estratificada por Efecto de la Música', fontsize=14)
plt.xlabel('Género Favorito', fontsize=12)
plt.ylabel('Edad Promedio', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```

Analisis:

-Todos los generos presentan por lo menos ciertas personas que mejoran su estado de salud mental, lo que nos indica que la salud mental esta directamente relacionada con el tipo de musica que se escucha.

A su vez, existen solo dos generos que no generan efectos en la salud mental, los demas si lo hacen, lo cual dicho con lo anterior, no bastaria solo con el grafico para poder realizar un analisis de cuanto influye el genero favorito de una persona en la salud mental, pero la conclusion obtenida es que si influye de manera positiva, neutra o negativa.

Entre alguno de los datos interesantes ademas de los ya encontrados, se tiene:


-Menos del 50% de los generos presentan efectos negativos en la salud mental entre los 10 y 30 años.

-La gente con mayor edad promedio presenta en casi todos los generos, mejoras o no efectos

Este grafico aunque no nos de toda la información necesaria, nos esta dando un indicio de similitud, al menos relacionando los BPM, entre el insomnio y OCD, y entre la ansiedad y la depresión

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data.csv'
data = pd.read_csv(file_path)

data_clean = data.dropna(subset=['Age','While working','Instrumentalist','Composer','Exploratory','Foreign languages','Music effects'])
data_clean['Age'] = pd.to_numeric(data_clean['Age'], errors='coerce')

data_clean['While working'] = data_clean['While working'].map({'Yes': 1, 'No': 0})
data_clean['Instrumentalist'] = data_clean['Instrumentalist'].map({'Yes': 1, 'No': 0})
data_clean['Composer'] = data_clean['Composer'].map({'Yes': 1, 'No': 0})
data_clean['Exploratory'] = data_clean['Exploratory'].map({'Yes': 1, 'No': 0})
data_clean['Foreign languages'] = data_clean['Foreign languages'].map({'Yes': 1, 'No': 0})
data_clean['Music effects'] = data_clean['Music effects'].map({'Improve': 1,'No effect':0.5,'Worsen': 0})
matriz = data_clean[['Age', 'While working', 'Instrumentalist', 'Composer', 'Exploratory', 'Foreign languages', 'Music effects']].corr()
sns.heatmap(matriz, annot=True)
plt.title("Mapa de calor para visualizar matriz de correlación 1")
plt.show()

data_clean_2 = data.dropna(subset=['Anxiety', 'Depression', 'Insomnia', 'OCD'])
data_clean_2['Depression'] = pd.to_numeric(data_clean['Depression'], errors='coerce')
data_clean_2['Anxiety'] = pd.to_numeric(data_clean['Anxiety'], errors='coerce')
data_clean_2['Insomnia'] = pd.to_numeric(data_clean['Insomnia'], errors='coerce')
data_clean_2['OCD'] = pd.to_numeric(data_clean['OCD'], errors='coerce')

data_clean_2['Depression'] = data_clean_2['Depression'].map({10: 1, 9: 0.9, 8: 0.8, 7: 0.7, 6: 0.6, 5: 0.5, 4: 0.4, 3: 0.3, 2: 0.2, 1: 0.1, 0: 0})
data_clean_2['Anxiety'] = data_clean_2['Anxiety'].map({10: 1, 9: 0.9, 8: 0.8, 7: 0.7, 6: 0.6, 5: 0.5, 4: 0.4, 3: 0.3, 2: 0.2, 1: 0.1, 0: 0})
data_clean_2['Insomnia'] = data_clean_2['Insomnia'].map({10: 1, 9: 0.9, 8: 0.8, 7: 0.7, 6: 0.6, 5: 0.5, 4: 0.4, 3: 0.3, 2: 0.2, 1: 0.1, 0: 0})
data_clean_2['OCD'] = data_clean_2['OCD'].map({10: 1, 9: 0.9, 8: 0.8, 7: 0.7, 6: 0.6, 5: 0.5, 4: 0.4, 3: 0.3, 2: 0.2, 1: 0.1, 0: 0})

matriz_2 = data_clean[['Anxiety', 'Depression', 'Insomnia', 'OCD']].corr()
sns.heatmap(matriz_2, annot=True)
plt.title("Mapa de calor para visualizar matriz de correlación 2")
plt.show()
```

En los resultados se evidencia una matriz de correlación que nos indica cuanto estan relacionados un topico con el otro siendo 1, la relacion perfecta es decir si uno aumenta el otro aumenta, y siendo -1 si uno aumenta el otro disminuye, el analisis de los resultados se hara por casos.

Respecto a la matriz de correlación 1:

-Si una persona toca musica, se espera que exista una relación evidente entre si la compone, lo cual se presenta como la relación mas fuerte de la matriz de correlacion.

-La otra relación destacable es que si el encuestado explora nuevos musicas, hable otros idiomas, significa que le gusta explorar todo tipo de musicas.

-Teniendo esas dos consideraciones, tanto exploratory,while working, instrumentalist, foreign languages y music effects(excluyendo la relacion entre estos ultimos dos) presentan relaciones con un mayor numero de correlación, mientras que instrumentalist y composer se presentan como variables mas aisladas entre si.


Respecto a la matriz de correlación 2:

-Como se esperaba existe una relación muy fuerte entre la depresión y la ansiedad.

-La ansiedad presenta la mayor relación entre los otros tres indicadores de salud mental, presentando la unica fila que tiene una correlacion sobre 0.25.

-La ansiedad con OCD y la depresion con Insomnia tambien presentan un grado de correlación alto, señalando que estan relacionadas tambien.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

file_path = 'data.csv'
data = pd.read_csv(file_path)

data_clean = data.dropna(subset=['Hours per day', 'Age'])
data_clean['Age'] = pd.to_numeric(data_clean['Age'], errors='coerce')



model = LinearRegression()
model.fit(data_clean[['Age']], data_clean['Hours per day'])


y_predict = model.predict(data_clean[['Age']])

plt.figure(figsize=(10, 6))
plt.scatter(x=data_clean['Age'], y=data_clean['Hours per day'], label = "Datos", color = "blue")
plt.plot(data_clean['Age'], y_predict, label = "Línea de regresión", color = "red")
plt.xlabel("Age")
plt.ylabel("Horas por dia")
plt.title("Modelo de regresión lineal")
plt.legend()
plt.show()


```
